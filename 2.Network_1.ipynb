```python
import pandas as pd
df = pd.read_csv('hospital.csv')
```

```python
df = df.fillna(999)
df['Scrub ID'] = df['Scrub ID'].astype('int')
df['First Assistant ID'] = df['First Assistant ID'].astype('int')
df['Second Assistant ID'] = df['Second Assistant ID'].astype('int')
df = df.astype('str')
```

```python
df['sur'] = 'sur'
df['ane'] = 'ane'
df['scr'] = 'scr'
df['cir'] = 'cir'
df['fai'] = 'fai'
df['sai'] = 'sai'
df['Surgeon ID'] = df['sur'].str.cat(df['Surgeon ID'])
df['Anesthesiologist ID'] = df['ane'].str.cat(df['Anesthesiologist ID'])
df['Scrub ID'] = df['scr'].str.cat(df['Scrub ID'])
df['Circulator ID'] = df['cir'].str.cat(df['Circulator ID'])
df['First Assistant ID'] = df['fai'].str.cat(df['First Assistant ID'])
df['Second Assistant ID'] = df['sai'].str.cat(df['Second Assistant ID'])
```

```python
df['SSC ID'] = df['Surgeon ID'].str.cat(df['Scrub ID'],sep = ',')
df['SC ID'] = df['Surgeon ID'].str.cat(df['Circulator ID'],sep = ',')
df['SF ID'] = df['Surgeon ID'].str.cat(df['First Assistant ID'],sep = ',')
df['SSA ID'] = df['Surgeon ID'].str.cat(df['Second Assistant ID'],sep = ',')
df['ASC ID'] = df['Anesthesiologist ID'].str.cat(df['Scrub ID'],sep = ',')
df['AC ID'] = df['Anesthesiologist ID'].str.cat(df['Circulator ID'],sep = ',')
df['SF ID'] = df['Anesthesiologist ID'].str.cat(df['First Assistant ID'],sep = ',')
df['ASA ID'] = df['Anesthesiologist ID'].str.cat(df['Second Assistant ID'],sep = ',')
df['SCC ID'] = df['Scrub ID'].str.cat(df['Circulator ID'],sep = ',')
df['SCF ID'] = df['Scrub ID'].str.cat(df['First Assistant ID'],sep = ',')
df['SCSA ID'] = df['Scrub ID'].str.cat(df['Second Assistant ID'],sep = ',')
df['CF ID'] = df['Circulator ID'].str.cat(df['First Assistant ID'],sep = ',')
df['CA ID'] = df['Circulator ID'].str.cat(df['Anesthesiologist ID'],sep = ',')
df['CSA ID'] = df['First Assistant ID'].str.cat(df['Second Assistant ID'],sep = ',')
```
